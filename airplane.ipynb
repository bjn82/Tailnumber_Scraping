```python
#value counts 
final_airplane['Manufacturer'].value_counts()
```

```python
#get value counts of columns
print('Type Aircraft:\n',final_airplane['Type_Aircraft'].value_counts())
print('\nType Engine:\n',final_airplane['Type_Engine'].value_counts())
print('\nEngine Model:\n',final_airplane['Engine_Model'].value_counts())
print('\nAirworthiness Category:\n',final_airplane['Airworthiness_Category'].value_counts())
print('\nAirworthiness Classification:\n',final_airplane['Airworthiness_Classification'].value_counts())
```

```python
#Filter out rows that dont fit in with what we are looking at
a = final_airplane.groupby("Engine_Model").filter(lambda x: len(x) >=20)
b = final_airplane.groupby("Airworthiness_Category").filter(lambda x: len(x) >=9)
c = final_airplane.groupby("Airworthiness_Classification").filter(lambda x: len(x) >=20)
d = final_airplane.groupby("Type_Aircraft").filter(lambda x: len(x) >=2)
```

```python
#merge all filtered dataframes back together only on matching tailnums
finalfinalairplane = pd.merge(a, b,  how = 'inner')
finalfinalairplane = pd.merge(finalfinalairplane, c,  how = 'inner')
finalfinalairplane = pd.merge(finalfinalairplane, d,  how = 'inner')
```

```python
#get value counts of columns
print('Type Aircraft:\n',finalfinalairplane['Type_Aircraft'].value_counts())
print('\nType Engine:\n',finalfinalairplane['Type_Engine'].value_counts())
print('\nEngine Model:\n',finalfinalairplane['Engine_Model'].value_counts())
print('\nAirworthiness Category:\n',finalfinalairplane['Airworthiness_Category'].value_counts())
print('\nAirworthiness Classification:\n',finalfinalairplane['Airworthiness_Classification'].value_counts())
```

```python
#Filter out values that arent needed
finalfinalairplane = finalfinalairplane[(finalfinalairplane.Airworthiness_Classification == 'Standard')|
                                        (finalfinalairplane.Airworthiness_Classification == 'Experimental')]

finalfinalairplane = finalfinalairplane[(finalfinalairplane.Airworthiness_Category == 'Transport')]
                                        
finalfinalairplane = finalfinalairplane[(finalfinalairplane.Engine_Model != 'UNKNOWN ENG')]
finalfinalairplane = finalfinalairplane[(finalfinalairplane.Engine_Model != 'Unknown')]
                        
```
